### Missing values in Test dataset
Before encoding, lets fill the missing values in the test dataset for the above columns.

```python
cols_to_fill = test.isnull().sum()[test.isnull().sum() > 0].index
```

```python
def fill_missing(df, col_name):
    if df[col_name].dtypes in [np.int, np.float]:
        df[col_name].fillna(value = train[col_name].median(), inplace=True)
    elif df[col_name].dtypes == np.object:
        df[col_name].fillna(value = train[col_name].mode()[0], inplace=True)
    return df

#is there any way I can do this better using apply or applymap?
for col in cols_to_fill:
    test = fill_missing(test, col)
```

#### Encoding catergorical variables

Based on our exploratory data analysis, we identified the following catergorical columns as important - 
`MSZoning, Neighborhood, LotShape, MasVnrType, ExternalQual, KitchenQual, HeatingQC, GarageFinish,  BsmtQual, CentralAir, LandContour, LotConfig, LandSlope, PavedDrive, RoofStyle`.

Lets figure out how to encode these categorical columns. 

Label Encoding: Converts each value in a column to a number. Not a good option for nominal (where values don't have an inherent order) variables.
ExternalQual, KitchenQual, HeatingQC, BsmtQual, OverallQual, GarageCars, TotalBaths

One Hot Encoding: Converts each category value into a new column with 1/0 denoting the presence or absence of the value for a particular observation.
MSZoning, Neighborhood, LotShape, MasVnrType, GarageFinish, LandContour, LotConfig, RoofStyle

LandSlope, PaveDrive

MS Zoning

```python
#create dummy variables for 
#ExternalQual, KitchenQual, HeatingQC, BsmtQual, 
#label based columns already exist for OverallQual, GarageCars, TotalBaths

map_ktch_exter_vals = {
    'Fa': 2,
    'TA': 3,
    'Gd': 4,
    'Ex': 5
}

map_bsmt_vals = {
    'None': 0,
    'Fa': 2,
    'TA': 3,
    'Gd': 4,
    'Ex': 5
}

map_heat_vals = {
    'Po': 1,
    'Fa': 2,
    'TA': 3,
    'Gd': 4,
    'Ex': 5
}


def encode_cols(df):
    df['ExterQual_enc'] = df['ExterQual'].replace(map_ktch_exter_vals)
    df['KitchenQual_enc'] = df['KitchenQual'].replace(map_ktch_exter_vals)
    df['BsmtQual_enc'] = df['BsmtQual'].replace(map_bsmt_vals)
    df['HeatingQC_enc'] = df['HeatingQC'].replace(map_heat_vals)
    return df


train = encode_cols(train)
test = encode_cols(test)
```

```python
#finally create dummy variables for 
#MSZoning, Neighborhood, LotShape, MasVnrType, GarageFinish

one_hot_enc_cols = ['MSZoning', 'Regular_LotShape', 'MasVnrType', 'GarageFinish', 'CentralAir', 'PavedDrive', 
                    'Gtl_LandSlope', 'LandContour', 'LotConfig', 'RoofStyle', 'Neighborhood']
def dummy_encode(df):
    df = pd.get_dummies(df[one_hot_enc_cols], 
                        prefix = one_hot_enc_cols)
    return df


train = pd.concat([train, dummy_encode(train)], axis=1)
test = pd.concat([test, dummy_encode(test)], axis=1)
```

### New Features

Lets check if we can create any new features from the current ones.

```python
numerical_cols
```

```python
cat_cols
```

1. We can use `YearRemodAdd` and `YrSold` columns to calculate the effective age of the house when it was sold. 
2. Lets combine the bath columns in the test and train dataset to get one column with total number of bathrooms.
3. We can combine `TotalBsmtSF, 1stFloorSF, 2ndFloorSF` to get a new column - `TotalSF`
4. Using the `Fireplaces` column, create a new binary column `HasFireplace` where a value of 0 indicates the absence of fireplace and 1 indicates a presence of fireplace
5. Create a `Has2ndFlr` column from `2ndFloorSF`

```python
def add_new_cols(df):
    df['EffHouseAge'] = df['YrSold'] - df['YearRemodAdd']
    df['TotalBaths'] = df['BsmtFullBath']  + df['FullBath'] + ((df['HalfBath'] + df['BsmtHalfBath'])*0.5)
    df['TotalSF'] = df['TotalBsmtSF'] + df['1stFlrSF'] + df['2ndFlrSF']
    df['HasFireplace'] = df['Fireplaces'].apply(lambda x: 0 if x == 0 else 1)
    df['Has2ndFlr'] = df['2ndFlrSF'].apply(lambda x: 0 if x == 0 else 1)
    return df

train = add_new_cols(train)
test = add_new_cols(test)
```
